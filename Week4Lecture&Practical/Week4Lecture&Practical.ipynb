```python
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0)
```

```python
df.head()
```

Now step through the examples in the tutorial to learn a little about the plot methods available in Seaborn.  

```python
# Seaborn plotting Function

sns.lmplot(x = "Attack",y = "Defense", data = df)
sns.set_style("darkgrid")
```

```python
# Seaborn doesnt have a dedicated scatter plot function. which is why you see a diagonal line.
# We actually used Seaborns function for fitting and plotting a regression line.

# Here we will remove the regression line and make the points colorfull as pockemon evolution

sns.set_style("darkgrid")
sns.lmplot(data = df, x= "Attack", y = "Defense",fit_reg = False, hue = "Stage") # lmplot means linearegression model(lm)
# ekhane fit_reg = False mane j regression line chilo oita remove kora hoise
# hue = "Stage" ekhane Stage name er ekta column ase oi column ta k colorfull kora hoise with different colors
```

```python
# Customize with Matplotlib
# If we wish to expand our axes in both direction we may need to use matplotlib

sns.set_style("darkgrid")
sns.lmplot(data = df, x = "Attack", y = "Defense",fit_reg = False,hue = "Stage")

#Matplotlib
plt.xlim(0,160) # xlim means x axis limitation
plt.ylim(0,200) # ylim means y axis limitation
```

```python
# Pandas role in Seaborn
# in boxplot we need to set the figure size using pandas 

plt.figure(figsize = (8,5)) # set the dimension of the plot by pandas

# There are 14 Data Visualization Plots of Seaborn
sns.boxplot(data = df)
```

```python
# Now we are going to create a new dataframe with more compitable columns or datasets
# Here we can see Total, Stage and Legendary Columns are not compitable with others.So we are going to drop those columns

stat_df = df.drop(["Total","Stage","Legendary"], axis = "columns")
# ekhane column theke drop kortesi tai mention kore disi kon axis theke drop hobe
# axis = "column" or axis = 1 for x axis or column dropping
# axis = "rows" or axis = 0 for y axis or row dropping
```

```python
# here is the new data set after dropping 

plt.figure(figsize = (8,5))
sns.boxplot(data = stat_df)
```

```python
# Seaborn theme
# There are 5 preset themes, and the default one is called darkgrid

plt.figure(figsize = (10,5))
sns.set_style("whitegrid")

# violinplot

# we can visualize the distribution of Attack by Pokmons primary type
sns.violinplot(data = df, x = "Type 1", y = "Attack")
```

Multivariate_normal

The multivariate normal, multinormal or Gaussian distribution is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and covariance matrix. These parameters are analogous to the mean (average or center) and variance (standard deviation, or width, squared) of the one-dimensional normal distribution.

```python
#Matplotlib Limitations
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed()
sample1 = np.random.multivariate_normal([0,0], [[1,0],[0,1]], 100) #Draw random samples from a multivariate normal distribution.
sample2 = np.random.multivariate_normal([1,1], [[0.05,0],[0,0.05]], 100)
print(sample1.shape, sample2.shape)
```

```python
sample = np.concatenate([sample1 , sample2])
sample.shape
```

```python
data_points_1 = plt.scatter(sample1[:,0],sample1[:,1],c = 'b')
data_points_2 = plt.scatter(sample2[:,0],sample2[:,1],c = 'r')

# plotting legends

plt.legend([data_points_1,data_points_2],['sample1','sample2'],loc= 'upper right')
```

```python
# Adding more points 
np.random.seed()
sample1 = np.random.multivariate_normal([0,0],[[0,1],[1,0]],5000)
sample2 = np.random.multivariate_normal([1,1],[[0.05,0],[0,0.05]],5000)
sample = np.concatenate([sample1,sample2])
plt.scatter(sample1[:,0],sample2[:,1])
```

```python
#  Creating a slider using bokeh

from bokeh.models import ColumnDataSource,CustomJS,Slider
from bokeh.layouts import column

x=[x*0.005 for x in range(0,201)]
source = ColumnDataSource(data=dict(x = x,y = x))

plot = figure(width=400, height=400)
plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)

slider = Slider(start = 0.10,end = 6, value = 1, step = .1, title = 'Power')

update_curve = CustomJS(args=dict(source = source, slider = slider),code="""

    var data = source.data;
    var f = slider.value;
    var x = data['x']
    var y = data['y']
    for (var i = 0; i < x.length; i++) {
        y[i] = Math.pow(x[i], f) 
    }
    
    source.change.emit();
""")
slider.js_on_change('value', update_curve)
                        
show(column(slider, plot))
                       
```
